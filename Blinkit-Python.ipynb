##### **Data Cleaning**

```python
print(df['Item Fat Content'].unique())
```

```python
df['Item Fat Content']=df['Item Fat Content'].replace({
    'reg':'Regular',
     'LF':'Low Fat',
    'low fat':'Low Fat'
})
```

##### **KPI Requirements**

```python
# Total Sales
Total_Sales=df['Sales'].sum()
#Average Sales
Average_Sales=df['Sales'].mean()
#No of items Sold
No_Of_Items_Sold=df['Sales'].count()
#Average Ratings
Average_Rating=df['Rating'].mean()
#Display
print(f"Total Sales:${Total_Sales:,.0f}")
print(f"Average_Sales:{Average_Sales:,.0f}")
print(f"No_Of_Items_Sold:{No_Of_Items_Sold:,.0f}")
print(f"Average_Rating:{Average_Rating:,.1f}")
```

##### **Chart's Requirements**

###### **Total Sales by fat content**

```python
sales_by_fat=df.groupby('Item Fat Content')['Sales'].sum()
plt.pie(sales_by_fat,autopct='%.1f%%', startangle=90,labels=sales_by_fat.index,shadow="true",explode=[0,0.1],colors=['#ffed29','#9acd32'])
f1={'family':'arial','size':15,'color':'orange'}
plt.title("Sales by fat content",fontdict=f1)
plt.show()
```

###### **Total Sales by Item Type**



```python
Sales_by_type=df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)
bars=plt.bar(Sales_by_type.index,Sales_by_type.values,width=0.9,color='#ffed29')
plt.xticks(rotation=-90)
f1={'family':'arial','size':15,'color':'orange'}
plt.title('Total Sales by Item Type',fontdict=f1)
plt.show()
```

##### **Fat content by outlet for total sales**


```python
grouped=df.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].sum().unstack()
grouped=grouped[['Regular','Low Fat']]
ax=grouped.plot(kind='bar',color=['#9acd32','#ffed29'])
f1={'family':'arial','size':15,'color':'orange'}
plt.title('Fat content by outlet for total sales',fontdict=f1)
f2={'family':'arial','size':10,'color':'Green'}
plt.xlabel('Outlet Location Type',fontdict=f2)
plt.ylabel('Total Sales',fontdict=f2)
plt.legend(title='Item Fat Content',loc='upper left',framealpha=0.5,facecolor='#FCFCFC',edgecolor='#F9F9F8',fancybox=True,shadow=True)
plt.tight_layout()
plt.show()
```

##### **toatal sales by outlet Establishment**


```python
sales_by_year=df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()
plt.plot(sales_by_year.index,sales_by_year.values,marker='o',linestyle='-',mfc='#ffed29',color='#9acd32')
f1={'family':'arial','size':15,'color':'orange'}
f2={'family':'arial','size':10,'color':'Green'}
plt.xlabel('Outlet establishment year',fontdict=f2)
plt.ylabel('Total sales',fontdict=f2)
plt.grid(axis='both',color='#97978f',ls='dotted')
plt.title('Outlet establishment',fontdict=f1)
plt.tight_layout()
plt.show()
```

##### **Total sales by outlet size**


```python
sales_by_size=df.groupby('Outlet Size')['Sales'].sum()
plt.pie(sales_by_size,labels=sales_by_size.index,autopct='%1.1f%%', startangle=90,colors=['orange','#9acd32','#ffed29'],explode=[0.05,0,0],shadow=True)
f1={'family':'arial','size':15,'color':'orange'}
plt.title('Total sales by Outlet size',fontdict=f1)
plt.show()
```

##### **Sales by outlet Location**


```python
sales_by_lacation=df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_lacation=sales_by_lacation.sort_values('Sales',ascending=False)
plt.figure(figsize=(10,2))
ax=sns.barplot(x='Sales',y='Outlet Location Type',data=sales_by_lacation,color='#9acd32')
f1={'family':'arial','size':15,'color':'orange'}
plt.title('Total sales by outlet location type',fontdict=f1)
f2={'family':'arial','size':10,'color':'Green'}
plt.xlabel('Total Sales',fontdict=f2)
plt.ylabel('Outlet Location Type',fontdict=f2)
plt.show()
```
